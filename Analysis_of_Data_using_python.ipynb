```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
import random
style.use("ggplot")
from google.colab import files
```

```python
#Create a NumPy array x with values ranging from 0 to 10.
x = np.arange(0,11)
x
```

```python
#Create a corresponding NumPy array y with values equal to the square of x.
y = [i**2 for i in x]
y
```

```python
#Plot y against x using a blue line
#Customize the plot with a title, xlabel, and ylabel.
plt.plot(x,y, c = "b")
plt.title("Squares values ranging From 0 to 10", fontsize="20")
plt.xlabel("Values Ranging From 0 to 10")
plt.ylabel("Squared Values")
plt.show()
```

```python
# Perform basic mathematical operations
# Add x and y
z = x + y

# Multiply x by a constant, let's choose 2 as an example
x2 = x * 2

# Plot the results of each operation
plt.figure(figsize=(14, 6))

# Plotting x + y
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot(x, z, label='x + sin(x)')
plt.xlabel('x')
plt.ylabel('x + sin(x)')
plt.title('Addition of x and sin(x)')
plt.legend()

# Plotting x multiplied by a constant
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.plot(x, x2, label='x * 2', color='red')
plt.xlabel('x')
plt.ylabel('x * 2')
plt.title('Multiplication of x by a constant')
plt.legend()

plt.tight_layout()
plt.show()
```

```python

# Sample data for plotting
x = np.arange(1, 5)
y = np.power(x, 2)  # y = x^2
z = np.sqrt(x)      # z = sqrt(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 23, 45, 12]
data = np.random.randn(100)  # Generating some random data for histogram

# Create a 2x2 subplot
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
axs[0, 0].plot(x, y, color='red', marker='o', linestyle='--', label='y = x^2')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('x axis')
axs[0, 0].set_ylabel('y axis')
axs[0, 0].legend()

# Scatter plot
axs[0, 1].scatter(x, z, color='blue', marker='^', label='z = sqrt(x)')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('x axis')
axs[0, 1].set_ylabel('z axis')
axs[0, 1].legend()

# Bar plot
axs[1, 0].bar(categories, values, color=['green', 'blue', 'orange', 'red'])
axs[1, 0].set_title('Bar Plot')
axs[1, 0].set_xlabel('Categories')
axs[1, 0].set_ylabel('Values')

# Adding text labels above bars
for i, v in enumerate(values):
    axs[1, 0].text(i, v + 1, str(v), ha='center', va='bottom')

# Histogram
axs[1, 1].hist(data, color='grey', edgecolor='black')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show plot
plt.show()
```

```python

# Step 1: Create a 1D NumPy array
arr_1d = np.arange(12)
print("1D Array:\n", arr_1d)

# Step 2: Reshape the 1D array into a 2D array with 3 rows and 4 columns
arr_2d = arr_1d.reshape(3, 4)
print("\n2D Array (3x4):\n", arr_2d)

# Step 3: Reshape the 2D array into a 3D array with 2 matrices, each having 3 rows and 2 columns
arr_3d = arr_2d.reshape(2, 3, 2)
print("\n3D Array (2 matrices, each 3x2):\n", arr_3d)
```

```python

# Task 11 arrays
arr_1d = np.arange(12)
arr_2d = arr_1d.reshape(3, 4)
arr_3d = arr_2d.reshape(2, 3, 2)

# Create figure and axes
fig, axs = plt.subplots(1, 4, figsize=(20, 5))  # 1 row, 4 columns for each array and the 3D array matrices

# Visualizing the 1D array (reshaped as 2D for visualization)
axs[0].imshow(arr_1d.reshape(1, 12), cmap='viridis')
axs[0].set_title('1D Array')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Value')
axs[0].set_xticks(range(12))

# Visualizing the 2D array
axs[1].imshow(arr_2d, cmap='viridis')
axs[1].set_title('2D Array (3x4)')
axs[1].set_xlabel('Column')
axs[1].set_ylabel('Row')

# Visualizing the first matrix of the 3D array
axs[2].imshow(arr_3d[0], cmap='viridis')
axs[2].set_title('3D Array Matrix 1')
axs[2].set_xlabel('Column')
axs[2].set_ylabel('Row')

# Visualizing the second matrix of the 3D array
axs[3].imshow(arr_3d[1], cmap='viridis')
axs[3].set_title('3D Array Matrix 2')
axs[3].set_xlabel('Column')
axs[3].set_ylabel('Row')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Create two 1D arrays of the same length
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Stack the arrays vertically
arr_vstack = np.vstack((arr1, arr2))
print("Vertically stacked:\n", arr_vstack)

# Stack the arrays horizontally
arr_hstack = np.hstack((arr1, arr2))
print("\nHorizontally stacked:\n", arr_hstack)

arr_concat_axis0 = np.concatenate((arr1.reshape(1, -1), arr2.reshape(1, -1)), axis=0)
print("\nConcatenated along axis 0 (equivalent to vstack):\n", arr_concat_axis0)


arr1_2d = arr1.reshape(1, -1)  # Reshape arr1 to 2D
arr2_2d = arr2.reshape(1, -1)  # Reshape arr2 to 2D
arr_concat_axis1 = np.concatenate((arr1_2d, arr2_2d), axis=1)
print("\nConcatenated along axis 1 (after reshaping to 2D, equivalent to hstack):\n", arr_concat_axis1)
```

```python
# Step 1: Load an image
# Replace 'image.png' with the path to your image file or URL
image_path = 'image.png'
image = plt.imread(image_path)

# Step 2: Display the original image and its shape
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')  # Hide axis
print("Original shape:", image.shape)

# Step 3: Reshape the image into a 1D array
image_1d = image.reshape(-1)
print("1D array shape:", image_1d.shape)

# Step 4: Reshape the 1D array back into the original shape
image_reshaped = image_1d.reshape(image.shape)

# Step 5: Display the reshaped image
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.imshow(image_reshaped)
plt.title('Reshaped Image')
plt.axis('off')  # Hide axis

plt.tight_layout()
plt.show()
```

```python

# Step 1: Create a 2D array
arr_2d = np.random.rand(3, 4)
print("Original array:\n", arr_2d)

# Step 2: Transpose the array
# You can use either np.transpose(arr_2d) or arr_2d.T
arr_2d_transposed = arr_2d.T
print("Transposed array:\n", arr_2d_transposed)

# Step 3: Visualize the original and transposed arrays
plt.figure(figsize=(12, 6))

# Visualizing the original array
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.imshow(arr_2d, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title('Original Array')

# Visualizing the transposed array
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.imshow(arr_2d_transposed, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title('Transposed Array')

plt.tight_layout()
plt.show()
```
